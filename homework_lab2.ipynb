```python
cr7_portugal_tbl = cr7_bs4.find_all('table','wikitable')[-2] 
```

```python
from IPython.core.display import HTML
HTML(str(cr7_portugal_tbl))
```

```python
rows = [row for row in cr7_portugal_tbl.find_all('tr')]
```

```python
rows[0]
```

As this is national team stats, we can discard the column name National team(only 1 value), i.e had it been club stats we could have extracted the team column, but in this case it is univariate, hence taking[1:] slice below

```python
column_names = [col.getText() for col in rows[0].find_all('th')[1:]]
column_names
```

```python
rows[-1]
```

```python
rows[1]
```

note that we need to remove first *"td"* element of 1st row(national team), and last row(aggregate info)

```python
values = [td.getText() for row in rows[1:-1] for td in row.find_all('td') if not td.get('rowspan') ]
```

```python
values
```

```python
type(rows)
len(column_names)
values[1::len(column_names)]
```

```python
stacked_values = [values[i::len(column_names)] for i in range(len(column_names))]
```

```python
list(zip(*stacked_values))
```

```python
stacked_rows = list(zip(*stacked_values))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(stacked_rows,columns=column_names)
```

```python
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df1 = df[["Year","Ratio"]]
```

```python
df.dtypes
```

```python
df1 = df1.astype('float')
df1.dtypes
```

```python
x = df1.Year
y = df1.Ratio
print(x)
plt.plot(x,y, 'o' ,alpha=.05)
plt.xlabel("Year")
plt.ylabel("Goal Ratio")
#df1.plot()
```
